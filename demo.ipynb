```python
import numpy as np
from stream3 import stream3
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
```

```python
# grid
[x,y,z] =  [np.linspace(-50,50,100) for _ in range(3)]
x3, y3, z3 = np.meshgrid(x, y, z, indexing='ij')
```

```python
#dipile
r =  np.sqrt(x3**2+y3**2+z3**2)
Bx0 = 3*x3*z3/r**5
By0 = 3*y3*z3/r**5
Bz0 = (3*z3**2-r**2)/r**5
```

```python
starting_points  = [
    np.array([51,52,49]),np.array([52,55,51]),
    np.array([51,49,49]),np.array([49,49,51])]
```

```python
ax = plt.axes(projection='3d')
for sp in starting_points:
    [xl,yl,zl] = stream3(Bx0,By0,Bz0,sp,maxstep=1000)
    ax.plot3D(xl,yl,zl)
```
